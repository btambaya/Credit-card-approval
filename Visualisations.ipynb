```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
# loading the updated dataset
vis = pd.read_csv("updated.csv")
```

```python
# plotting the correlation plot
corr_matrix = vis.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, 
            annot=True, 
            linewidths=0.5, 
            fmt= ".2f", 
            cmap="YlGnBu");
```

```python
# plotting a chart for outcome
vis.STATUS.value_counts().plot(kind="bar", color=["skyblue","salmon"])

plt.title("Outcome Frequency")
plt.xlabel("Approval Outcome")
plt.ylabel("Frequency")
```

```python
# plotting a chart for gender
vis.CODE_GENDER.value_counts().plot(kind="bar", color=["skyblue","salmon"])

plt.title("Gender Frequency")
plt.xlabel("Gender Type")
plt.ylabel("Frequency")
plt.legend(['0 Female', '1 Male'])
```

```python
# age distribution plot showing majority of the applicants are 20 to 68 years old
vis.AGE.plot.hist(figsize=(10,6), bins=20)
plt.xlabel('Age')
plt.title('Age Frequency')
plt.show()
```

```python
vis.YEARS_WORKED.plot.hist()
plt.xlabel('No of Years Employed')
plt.title('Years Worked Frequency')
plt.show()
```

```python
# This graph shows that, majority of applicatant's are married
plt.figure(figsize=(8,8))
sns.barplot(vis['NAME_FAMILY_STATUS'].value_counts().index, vis['NAME_FAMILY_STATUS'].value_counts().values)
plt.title('% of Applications submitted based on Family Status')
plt.legend(['1 Separated','2 Marrried', '3 Single/Not Married', '4 Widow', '5 Civil Marriage '])
plt.show()
```

```python
# This graph shows that, majority of applicatant's lives in House / Apartment
plt.figure(figsize=(8,6))
sns.barplot(vis['NAME_HOUSING_TYPE'].value_counts().index, vis['NAME_HOUSING_TYPE'].value_counts().values)
plt.title('% of Applications submitted based on Housing Type')
plt.legend(['1 House/Apartment','2 Municipal Apartment', '3 Co-op Apartment', '4 Rented Apartment', '5 Office Apartment', '6 With Parent'])
plt.show()
```

```python
# This graph shows that, majority of application are submitted by Female's
plt.figure(figsize = (8,8))
plt.pie(vis['CODE_GENDER'].value_counts(), labels=['Female', 'Male'], autopct='%1.2f%%')
plt.title('% of Applications submitted based on Gender')
plt.show()
```

```python
# This graph shows that, majority of application are approved for Female's
plt.figure(figsize = (8,8))
plt.pie(vis[vis['STATUS']==0]['CODE_GENDER'].value_counts(), labels=['Female', 'Male'], autopct='%1.2f%%')
plt.title('% of Applications Approved based on Gender')
plt.show()
```

```python
# This graph shows that, majority of applicatant's dont own a car
plt.figure(figsize = (8,8))
plt.pie(vis['FLAG_OWN_CAR'].value_counts(), labels=['No', 'Yes'], autopct='%1.2f%%')
plt.title('% of Applications submitted based on owning a Car')
plt.show()
```

```python
# This graph shows that, majority of applicatant's own a Real Estate property / House
plt.figure(figsize = (8,8))
plt.pie(vis['FLAG_OWN_REALTY'].value_counts(), labels=['Yes','No'], autopct='%1.2f%%')
plt.title('% of Applications submitted based on owning a Real estate property')
plt.show()
```

```python
# This graph shows that, majority of applicatant's don't have any children
plt.figure(figsize = (8,8))
plt.pie(vis['CNT_CHILDREN'].value_counts(), labels=vis['CNT_CHILDREN'].value_counts().index, autopct='%1.2f%%')
plt.title('% of Applications submitted based on Children count')
plt.legend()
plt.show()
```

```python
# This graph shows that, majority of applicatant's are working professional
plt.figure(figsize = (8,8))
plt.pie(vis['NAME_INCOME_TYPE'].value_counts(), labels=vis['NAME_INCOME_TYPE'].value_counts().index, autopct='%1.2f%%')
plt.title('% of Applications submitted based on Income Type')
plt.legend(["4 Working", "2 Commercial Associate", "5 Pensioner", "3 State Servant", "1 Student "])
plt.show() 
```

```python
# This graph shows that, majority of applicatant's completed the Secondary Education
plt.figure(figsize=(8,8))
plt.pie(vis['NAME_EDUCATION_TYPE'].value_counts(), labels=vis['NAME_EDUCATION_TYPE'].value_counts().index, autopct='%1.2f%%')
plt.title('% of Applications submitted based on Education')
plt.legend(["3 Secondary/ Secondary Special", " 5 Higher Education", "2 Incomeplete Higher", "4 Lower Secondary", "1 Academic Degree "])
plt.show()
```

```python
# This graph shows that, majority of applications are rejected if Total income & years of Employment is less
sns.scatterplot(vis['YEARS_WORKED'], vis['AMT_INCOME_TOTAL'], hue=vis['STATUS'])
plt.title('Scatter Plot')
plt.show()
```
